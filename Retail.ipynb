```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Retail_Dataset (2).csv')
```

```python
sns.set(style="whitegrid")
```

```python
df.head(5)
```

```python
# ---------- UNIVARIATE ANALYSIS (4 examples) ----------
# 1. Distribution of Unit Price
plt.figure(figsize=(6,4))
sns.histplot(df['unit_price'], kde=True, color='skyblue')
plt.title('Distribution of Unit Price')
plt.show()
```

```python
# 2. Customer Type Count
# 2. Customer Type Count (Fixed)
plt.figure(figsize=(5,4))
sns.countplot(
    x='customer_type',
    hue='customer_type',   # Added hue to avoid warning
    data=df,
    palette='pastel',
    legend=False           # Hide duplicate legend
)
plt.title('Customer Type Count')
plt.show()
```

```python
# 3. Footfall Distribution
plt.figure(figsize=(6,4))
sns.histplot(df['footfall'], kde=True, color='lightgreen')
plt.title('Distribution of Footfall')
plt.show()
```

```python
# 4. Payment Mode Frequency
plt.figure(figsize=(6,4))
sns.countplot(
    x='payment_mode',
    hue='payment_mode',      # Added to remove warning
    data=df,
    palette='muted',
    legend=False             # Hide duplicate legend
)
plt.title('Payment Mode Count')
plt.show()
```

```python
# ---------- BIVARIATE ANALYSIS (2 examples) ----------
# 1. Average Net Revenue by Payment Mode (Fixed)
plt.figure(figsize=(6,4))
sns.barplot(
    x='payment_mode',
    y='net_revenue',
    hue='payment_mode',    # Added to satisfy Seaborn
    data=df,
    palette='muted',
    estimator='mean',
    errorbar=None,
    legend=False           # Hide duplicate legend
)
plt.title('Average Net Revenue by Payment Mode')
plt.show()
```

```python
# 2. Average Units Sold by Store Region (Fixed)
plt.figure(figsize=(6,4))
sns.barplot(
    x='store_region',
    y='units_sold',
    hue='store_region',    # Added to satisfy Seaborn
    data=df,
    palette='pastel',
    estimator='mean',
    errorbar=None,
    legend=False
)
plt.title('Average Units Sold by Store Region')
plt.show()
```

```python

# ---------- MULTIVARIATE ANALYSIS (2 examples) ----------
# 1. Net Revenue by Payment Mode & Customer Type
plt.figure(figsize=(7,5))
sns.barplot(x='payment_mode', y='net_revenue', hue='customer_type', data=df, palette='Set2')
plt.title('Net Revenue by Payment Mode & Customer Type')
plt.show()
```

```python
# 2. Heatmap of Correlation Between Numerical Features
numeric_cols = ['unit_price', 'units_sold', 'total_sale', 'net_revenue', 'footfall', 'staff_on_duty']
corr = df[numeric_cols].corr()
plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Sales Metrics')
plt.show()
```
